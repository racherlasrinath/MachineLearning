```python
y=df['charges']
```

```python
x=df.drop(columns=['charges','children'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from category_encoders import TargetEncoder
te=TargetEncoder()
x_train=te.fit_transform(x_train,y_train)
x_test=te.transform(x_test,y_test)
```

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
x_train=m.fit_transform(x_train)
x_test=m.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
```

```python
y_pred=l.predict(x_test)
```

```python
y_pred
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge
alpha=0
rid=Ridge(alpha=alpha)
rid.fit(x_train,y_train)
```

```python
train_ridge=rid.score(x_train,y_train)
test_ridge=rid.score(x_test,y_test)
```

```python
train_ridge
```

```python
test_ridge
```

```python
y_ridge_pred=rid.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_ridge_pred)
print(f"Mean squared error: {mse:.2f}")
```

```python
from sklearn.linear_model import Lasso
l=Lasso(alpha=0)
l.fit(x_train,y_train)
```

```python
train_lasso_score=l.score(x_train,y_train)
test_lasso_score=l.score(x_test,y_test)
```

```python
train_lasso_score
```

```python
test_lasso_score
```

```python
lasso_pred_y=l.predict(x_test)
```

```python
mse = mean_squared_error(y_test, lasso_pred_y)
mse
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, lasso_pred_y)
```
